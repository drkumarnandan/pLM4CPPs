```python
import pandas as pd

# Load the Excel file
file_path = 'CPP_Final_non_redundant_sequences-For_Length_AAdistribution.xlsx'
df = pd.read_excel(file_path)

# Compute the length of each sequence
df['length'] = df['sequence'].apply(len)

# Define length ranges
bins = list(range(0, 75, 5))
labels = [f"{bins[i]}-{bins[i+1]}" for i in range(len(bins)-1)]

# Categorize sequences into length ranges
df['length_range'] = pd.cut(df['length'], bins=bins, labels=labels, right=False)

# Count sequences in each length range
length_range_counts = df['length_range'].value_counts().sort_index()

# Display the DataFrame with sequence lengths
print(df)

# Display the length range counts
print(length_range_counts)

# Save the DataFrame with lengths and the length range counts to new Excel files
output_file_path_df = 'CPP_Sequences_with_Lengths.xlsx'
output_file_path_counts = 'CPP_Sequence_Length_Range_Counts.xlsx'

df.to_excel(output_file_path_df, index=False)
length_range_counts.to_excel(output_file_path_counts, index=True, header=['count'])

print(f"Data with lengths saved to {output_file_path_df}")
print(f"Length range counts saved to {output_file_path_counts}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel files
cpp_counts_file_path = 'CPP_Sequence_Length_Range_Counts.xlsx'
non_cpp_counts_file_path = 'Non-CPP_Sequence_Length_Range_Counts.xlsx'

cpp_counts = pd.read_excel(cpp_counts_file_path, index_col=0)
non_cpp_counts = pd.read_excel(non_cpp_counts_file_path, index_col=0)

# Add a column to distinguish between CPPs and Non-CPPs
cpp_counts['Type'] = 'CPPs'
non_cpp_counts['Type'] = 'Non-CPPs'

# Combine the dataframes
combined_counts = pd.concat([cpp_counts, non_cpp_counts]).reset_index()

# Remove the length range 65-70
combined_counts = combined_counts[combined_counts['length_range'] != '65-70']

# Plotting
plt.figure(figsize=(7, 4))  # Adjusted figure size for better readability
sns.barplot(x='length_range', y='count', hue='Type', data=combined_counts, palette='viridis')

# Add titles and labels with bold and size 12 font
plt.xlabel('Length Range', fontsize=12, fontweight='bold')
plt.ylabel('Count of Sequences', fontsize=12, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=10, fontweight='bold')
plt.yticks(fontsize=10, fontweight='bold')

# Customize legend to remove the title and set font size and weight
plt.legend(title='', fontsize=12, prop={'weight': 'bold'})

# Adjust bottom margin to make sure all parts are visible in saved plot
plt.subplots_adjust(bottom=0.25)

# Save plot as PNG file
plt.savefig('peptide_length_distribution.png')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from collections import Counter

# Load the Excel file
file_path = 'CPP_Final_non_redundant_sequences-For_Length_AAdistribution.xlsx'
df = pd.read_excel(file_path)

# Function to count amino acid residues
def count_residues(sequence):
    return dict(Counter(sequence))

# Count residues for each sequence
df['residue_counts'] = df['sequence'].apply(count_residues)

# Display the DataFrame with residue counts
print(df)

# Save the DataFrame with residue counts to a new Excel file
output_file_path = 'CPP_Sequences_Residue_Counts.xlsx'
df.to_excel(output_file_path, index=False)

print(f"Residue counts saved to {output_file_path}")
```

```python
import pandas as pd
from collections import Counter

# Load the Excel file
file_path = 'CPP_Final_non_redundant_sequences-For_Length_AAdistribution.xlsx'
df = pd.read_excel(file_path)

# Function to count amino acid residues
def count_residues(sequence):
    return dict(Counter(sequence))

# Count residues for each sequence
df['residue_counts'] = df['sequence'].apply(count_residues)

# Aggregate residue counts across all sequences
total_residue_counts = Counter()
for counts in df['residue_counts']:
    total_residue_counts.update(counts)

# Convert the Counter to a DataFrame for better display (optional)
total_residue_counts_df = pd.DataFrame(total_residue_counts.items(), columns=['Residue', 'Total_Count'])

# Display the total residue counts
print(total_residue_counts_df)

# Save the total residue counts to a new Excel file
output_file_path = 'CPP_Sequences_Residue_Counts.xlsx'
total_residue_counts_df.to_excel(output_file_path, index=False)

print(f"Total residue counts saved to {output_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# Set font properties globally for all text elements
matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['font.weight'] = 'bold'

# Load total residue counts for CPP sequences
cpp_file_path = 'CPP_Sequences_Residue_Counts.xlsx'
df_cpp = pd.read_excel(cpp_file_path)

# Load total residue counts for non-CPP sequences
non_cpp_file_path = 'Non-CPP_Sequences_Residue_Counts.xlsx'
df_non_cpp = pd.read_excel(non_cpp_file_path)

# Merge dataframes on 'Residue' column, filling NaN values with 0
merged_df = pd.merge(df_cpp, df_non_cpp, on='Residue', suffixes=('_CPP', '_Non-CPP'), how='outer').fillna(0)

# Calculate total counts of CPP and non-CPP sequences
total_cpp_sequences = 23505
total_non_cpp_sequences = 84918

# Convert counts to percentages
merged_df['Percentage_CPP'] = (merged_df['Total_Count_CPP'] / total_cpp_sequences) * 100
merged_df['Percentage_Non-CPP'] = (merged_df['Total_Count_Non-CPP'] / total_non_cpp_sequences) * 100

# Plotting
plt.figure(figsize=(7, 4))

# Calculate bar width and position adjustment
bar_width = 0.4
bar_positions_cpp = range(len(merged_df))
bar_positions_non_cpp = [pos + bar_width for pos in bar_positions_cpp]

# Use Viridis colormap colors
cpp_color = plt.cm.viridis(0.2)  # Darker shade for CPPs
non_cpp_color = plt.cm.viridis(0.7)  # Darker shade for Non-CPPs

plt.bar(bar_positions_cpp, merged_df['Percentage_CPP'], width=bar_width, color=cpp_color, alpha=0.7, label='CPPs')
plt.bar(bar_positions_non_cpp, merged_df['Percentage_Non-CPP'], width=bar_width, color=non_cpp_color, alpha=0.7, label='Non-CPPs')

plt.xlabel('Residue', fontsize=14, fontname='Times New Roman', fontweight='bold')
plt.ylabel('Percentage (%)', fontsize=14, fontname='Times New Roman', fontweight='bold')
plt.xticks(bar_positions_cpp, merged_df['Residue'], rotation=45, fontname='Times New Roman', fontsize=13, fontweight='bold')
plt.yticks(fontname='Times New Roman', fontsize=13, fontweight='bold')
plt.legend(prop={'size': 12, 'family': 'Times New Roman', 'weight': 'bold'})

# Adjust bottom margin to make sure all parts are visible in saved plot
plt.subplots_adjust(bottom=0.25)

# Save plot as PNG file
plt.savefig('Occurrence of Residues in Percentage.png')

plt.tight_layout()
plt.show()
```
